```python
from pymatgen.ext.matproj import MPRester
import numpy as np
from pymatgen.core.operations import SymmOp
from IPython.display import Image
from IPython.display import display, Math, Latex
import numpy as np
with MPRester("Osul7wWQrx1czIDzJdH7") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-24504")
    
    #1ère opération de symétrie : plan miroir qui envoie (x,y,z) en (x,y,-z+(1/2))
    point1 = np.array([0.33, 0.67, 0.17])
    sym1 = SymmOp.from_xyz_string('x,y,-z+(1/2)').operate(point1)
    print('1ère opération de symétrie : plan miroir de vecteur normal [0, 0, 1] qui envoie (x,y,z) en (x,y,-z+(1/2))')
    print(sym1)
    Im=Image(filename='Img1.PNG')
    display(Im)
    
    #2ème opération de symétrie : rotation C3 autour du vecteur [0, 0, 1] qui envoie (x,y,z) en (-y,x-y,z)
    point2 = np.array([0.67, 0.33, 0.83])
    sym2 = SymmOp.from_xyz_string('-y,x-y,z').operate(point2)
    print('2ème opération de symétrie : rotation C3 autour du vecteur [0, 0, 1] qui envoie (x,y,z) en (-y,x-y,z)')
    print(sym2)
    Im2=Image(filename='Img2.PNG')
    display(Im2)
    
    #3ème opération de symétrie : plan miroir qui envoie (x,y,z) en (x,y,-z+(1/2) 
    point3 = np.array([0.67, 0.33, 0.67])
    sym3 = SymmOp.from_xyz_string('x,y,-z+(1/2').operate(point3)
    print('3ème opération de symétrie : plan miroir qui envoie (x,y,z) en (x,y,-z+(1/2)')
    print(sym3)
    Im3=Image(filename='Img3.PNG')
    display(Im3)
```
