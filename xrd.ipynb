```python
from pymatgen.ext.matproj import MPRester
from IPython.display import Image
from IPython.display import display
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from tabulate import tabulate
with MPRester("Osul7wWQrx1czIDzJdH7") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-24504")
    
c = XRDCalculator()
print("On peut remarquer les diff√©rents pics sur cette figure")
c.show_plot(structure)
sp = c.get_pattern(structure)
# 3 premiers pics 
p1 = sp.hkls[0][0]
p2 = sp.hkls[1][0] 
p3 = sp.hkls[2][0]
print("Voici ensuite les caract√©ristiques des 3 premiers pics")
tab = [["Premier pic", p1["hkl"],sp.d_hkls[0], sp.x[0], sp.y[0]],["Deuxi√®me pic", p2["hkl"],sp.d_hkls[1], sp.x[1], sp.y[1]],["Troisi√®me pic", p3["hkl"],sp.d_hkls[2], sp.x[2], sp.y[2]],]  
print(tabulate(tab, headers=["","indices hkls", "d","theta", "Intensit√©"],stralign="center", numalign="center"),"\n")
```
